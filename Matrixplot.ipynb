```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

 # Heat map


```python
flights=sns.load_dataset("flights")
flights
```

```python
tips=sns.load_dataset("tips")
tips
```

```python
tipscorr=tips[['total_bill','tip','size']]
tipscorr
```

```python
tipscorr.corr()
```

```python
sns.heatmap(tipscorr.corr(),annot=True)
```

  # Cluster map


```python
sns.clustermap(tipscorr.corr(),annot=True)
```

 
 # Pivot table Heat map

```python
pvtflight=flights.pivot_table(values='passengers',index='month',columns='year')
pvtflight
```

```python
sns.heatmap(pvtflight,annot=True)
```
