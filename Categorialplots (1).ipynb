# Categorical Data Plots 
Now lets discuss using seaborn to plot categorical data! These are a few main plot types for this 
* box plot
* violinplot
* stripplot
* swarmplot
* barplot
* countplot
lets go through examples of each!


```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
df
```

```python
sns.countplot(x=df["sex"],hue=df['smoker'])
plt.title("Smokers")
plt.show()
```

```python
sns.barplot(x=df["sex"],y=df["total_bill"],estimator=np.sum)
```

```python
sns.boxplot(x='tip',y='day',data=df)
```

```python
import seaborn as sns

df=sns.load_dataset("tips")
df
```

```python
sns.pairplot(df,hue="sex")
```

```python
sns.countplot(x=df['sex'],hue=df['smoker'])
```

```python
sns.boxplot(x='tip',y='day',data=df,palette='rainbow')
```

violinplot

```python
sns.violinplot(x='day',y='total_bill',data=df)
```

# WHAT IS STIRPPLOT AND SWARM PLOT 
The stripplot will draw a scatterplot where one  variable is categorical . A strip plot can be drown as its own ,   but it is also a good compelment to a box or violin plot in cases where you eant yo show all obersvation along with some representation of the underlying distribution 

The swarmplot is similar to stripplot(), but the point are adjusted (only along the categorical axis) so that they don't overlap this gives better representation of the distribution of values, although it does not scales as well to large number of observation ( both in terms oif the ability to show all the point and in terms of the computation needed to arrange them)

```python
sns.stripplot(x='day',y='total_bill',data=df)
```

```python
sns.swarmplot(x='day',y='total_bill',data=df)
```

```python
sns.violinplot(x='tip',y='day',data=df)
```

```python
sns.violinplot(x='tip',y='day',data=df)
sns.swarmplot(x='tip',y='day',data=df)
```
