```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
x=np.linspace(0,5,11)
x
y=x**2
y
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x,y)
```

# object oriented way

```python
fig=plt.figure(figsize=(12,7),dpi=500)
axis1=fig.add_axes([0.8,0.8,0.4,0.6])
axis1.plot(x,y)
axis2=fig.add_axes([0.3,0.9,0.2,0.4])
axis2.plot(x,y)
```

# Types of plots

# Saving Plots

```python
fig=plt.figure()
a=fig.add_axes([0.1,0.1,0.8,0.8])
a.plot(x,y)
plt.savefig("basicplot.png")
```
