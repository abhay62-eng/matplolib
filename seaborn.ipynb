# We use distribution plots to visualize the distribution of quantative data.
# let discuss some plots that allow us to visualize the distribution of a data set . these plots are 
# displot/hisplot
# .joinpoint
# . pairplot
# . rug plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

```python
plt.subplot(1,2,1)
sns.histplot(df["total_bill"],bins=20,kde=True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins=20,kde=True)
```

# Joint plot

```python
sns.jointplot(x= 'total_bill', y= 'tip', data= df,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y="tip",data=df,kind='kde')
```

```python
sns.jointplot(x='total_bill',y="tip",data=df,kind='hex')
```

```python
sns.jointplot(x='total_bill',y="tip",data=df,kind='reg')
```

# Pair plot

```python
sns.pairplot(df,hue="size",palette='rainbow')
```

# rugplot

```python
sns.rugplot(df)
```

# Categorical Data plots 
now lets discuss using seaborn to plot categorical data There sare a few main plot types for this 
* boxplot
* violinplot
* stripplot
* swarmplot
* barplot
* countplot
lets go through exampls of each!


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
df
```

```python
# which is more males or female
sns.countplot(x=df['sex'],hue=df['smoker'])
```

```python
import numpy as np
sns.barplot(y=df['tip'],x=df['sex'],estimator=np.sum)
```

```python
# we using the boxplot in the categorical 
sns.boxplot(x='tip',y='day',data=df,palette='rainbow')
```
