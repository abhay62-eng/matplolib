# Trying to create multiple subplots in one figure
we are understand how to create a multiple subplots in one figure that each subplot will display a different dataset . this technique is useful when we want to compare multiple datasets side by side in a single figure .

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


a=[1,2,3,4,5]
b=np.arange(3)
c=[4,2,6,8,9,20,13,18]
fig=plt.figure(figsize=(10,10))
```

```python
sub1=plt.subplot(2,2,1)
sub2=plt.subplot(2,2,2)
sub3=plt.subplot(2,2,3)
sub4=plt.subplot(2,2,4)
sub1.plot(a,"sb")
sub1.set_title("1st Rep")

sub2.plot(b,"or")
sub2.set_xticks(list(range(0,10,12)))
sub2.set_title("2nd Rep")

sub3.plot(list(range(0,22,3)),"vg")
sub3.set_title("3rd Rep")
sub3.set_xticks(list(range(0,10,1)))

sub4.plot(c,"Dm")
sub4.set_yticks(list(range(0,24,2)))
sub4.set_title("4th Rep")
```

#  Plotting a scatter plot
scatter plots are basic type of plot used to visualize the relationship between two variable. This is usefull when we want to see how one variable is related to another or look for correlation.

```python
import matplotlib.pyplot as plt 
s=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.scatter(s,y,color="Blue",marker=".")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("simple scatter plot")
plt.grid(True)
plt.show()
```
